# Kiểu dữ liệu thời gian trong Python

## Giới thiệu
Khi nhắc đến thời gian, người ta sẽ thường đề cập đến 3 dạng sau :
- Thời điểm : là một điểm cụ thể trong dòng thời gian. Ví dụ như `16/01/2020`, `11/11/2019`, ...
- Khoảng thời điểm : là một khoảng trong dòng thời gian, có điểm đầu điểm cuối xác định. Ví dụ `năm 2019`, `quý 2 năm 2020`, ...
- Khoảng thời gian : chỉ độ dài của một khoảng trong dòng thời gian. Ví dụ : `86400 giây`, `1 năm 2 tháng 20 ngày`, ...

Sau đây, chúng ta cùng tìm hiểu một số cách phổ biến để làm việc với thời gian trong `python`

## Thư viện `datetime` và `dateutil`
Thư viện `datetime` và `dateutil` là hai thư viện đi kèm khi cài dặt python. Hai thư viện này cung cấp rất nhiều công cụ để làm việc với thời gian.

Ví dụ như bạn có thể tạo ra một thời điểm bằng đối tượng `datetime` của thư viện `datetime`

````python
from datetime import datetime

# tạo ra thời điểm
d1 = datetime(year = 2019, month = 11, day = 18, hour = 12, minute = 34, second = 56)
d1
````

````python
# hoặc lấy thời điểm hiện tại
d2 = datetime.now()
d2
````

Bạn cũng đó thể dịch chuỗi ngày tháng thành kiểu thời gian bằng thư viện `dateutil`

````python
from dateutil import parser

d3 = parser.parse('November 20th, 2019')
d3
````

Khi đã có đối tượng `datetime`, bạn co thể làm được nhiều điều như in ra thứ của đối tượng `datetime` đó bằng phương thức `.strftime()`

````python
d1.strftime('%A')
````

Phương thức `.strftime(<format>)` hỗ trợ chuyển đổi đối tượng `datetime` thành `str` theo định dang truyền vào. Quy ước về định dạng bạn có thể tham khảo tại đây <a href = "http://strftime.org/">đây</a>

`datetime` có ưu điểm là khá đơn giản nhưng mạnh mẽ, bạn có thể làm gần như mọi thứ bằng các đối tượng dựng sẵn của `datetime` và `dateutil`.

Tuy nhiên, khi làm việc với mảng thời gian lớn thì các kiểu đối tượng này không được hiệu quả lắm.

Vì vậy, chúng ta cùng đến với thư viện `numpy`.

## Thư viện `numpy`
Thư viện `numpy` hỗ trợ lưu trữ thời gian bằng đối tượng `datetime64`.

Để khởi tạo một thời điểm, ta có thể làm như sau :

````python
import numpy as np

# tạo thời điểm, chính xác đến giây
d4 = np.datetime64('2019-11-19', 's')
d4
````

Để tạo ra một mảng thời gian, chúng ta làm như sau

````python
# để tạo ra mảng thời điểm trong numpy, phải chỉ rõ dtype là một kiểu thời gian mà numpy hỗ trợ
da1 = np.array(['2019-10-01', '2019-11-03'], dtype = np.datetime64)
da1
````

Bạn cũng có thể sinh ra một mảng thời gian bằng `np.arange()` như sau :

````python
# giống như trên, để tạo mảng thời điểm bằng np.arange(), cũng phải chỉ ra dtype
da2 = np.arange('2019-11', '2019-12', dtype = 'datetime64[D]')
da2
````

Tuy nhiên, đối tượng `numpy.datetime64` lại không có phương thức tương tự `.strftime()`

Các bạn có thể đọc thêm về đối tượng `numpy.datetime64` tại <a href = "https://docs.scipy.org/doc/numpy/reference/arrays.datetime.html">đây</a>

## Các thao tác trên `Time Series`
Về cơ bản, các thao tác trên `Series` đều có thể thực hiện trên `Time Series`. 

Tuy nhiên, `Time Series` cũng có những điểm riêng của mình.

### Truy xuất dữ liệu

Với `Time Series`, bạn có thể truy xuất bằng khoảng thời điểm

````python
# tạo dữ liệu ngẫu nhiên
data_1 = np.random.randint(10000, size = 365)
# tạo DatetimeIndex
index_1 = pd.date_range(start = '2019-01-01', periods = 365, freq = 'D')
# tạo Time Series
ts1 = pd.Series(data = data_1, index = index_1)
ts1
````

````python
# lấy dữ liệu trong tháng 05 năm 2019
ts1['2019-05']
````

Bạn có thể truy xuất bằng phương thức `.asfreq()` như sau :
```
<tên_Time_Series>.asfreq(<tên_phương_pháp>)
```
Trong đó, `<tên_phương_pháp>` được quy ước như sau :

|Tên phương thức|Ý nghĩa||Tên phương thức|Ý nghĩa|
|:---:|---|---|:---:|---|
|`D`|Ngày trong năm||`B`|Ngày làm việc|
|`W`|Cách tuần|
|`M`|Ngày cuối tháng||`MS`|Ngày đầu tháng|
|`Q`|Ngày cuối quý||`QS`|Ngày đầu quý|
|`A`|Ngày cuối năm||`AS`|Ngày đầu năm|

Nói thêm :
- `W` sẽ lấy dữ liệu là của các ngày chủ nhật, nếu muốn đổi thứ khác, có thể dùng `W-MON`, `W-TUE`, ..., `W-SAT`
- Bạn có thể thay đổi cách tính của một quý bằng cách chỉ rõ tháng bắt đầu (kết thúc). Ví dụ `Q-FEB` nghĩa là ngày cuối của quý là ngày cuối tháng hai. Như vậy, `Q` = `Q-MAR` 

````python
# lấy dữ liệu của ngày đầu tháng
ts1.asfreq('MS')
````

### Thêm dữ liệu mới
Tương tự như `Series`, bạn có thể thêm dữ liệu mới bằng cú pháp :
```
<tên_Time_Series>.[<label>] = <dữ_liệu>
```
Tuy nhiên, cần lưu ý là `<label>` phải là `Timestamp`.

````python
# tạo Time Series
data_2 = [1, 2, 3]
index_2 = pd.to_datetime(['2019-01-01', '2019-02-01', '2019-03-01'])
ts2 = pd.Series(data = data_2, index = index_2)
ts2
````

````python
# thêm dữ liệu mới
ts2[pd.Timestamp('2019-04-01')] = 4
ts2
````

### Gộp dữ liệu

`Time Series` còn cho phép bạn gộp dữ liệu lại trước khi thực hiện các thao tác thông kê bằng phương thức `.resample()`.

Cú pháp :
```
<tên_Time_Series>.resample(<cách_gộp>)
```
Một số cách gộp đơn giản như :
- `W` : gộp theo tuần
- `M` : gộp theo tháng
- `A` : gộp theo năm

````python
# tính tổng theo tháng
ts1.resample('M').sum()
````

````python
# tính trung bình theo quý
ts1.resample('Q').mean()
````
